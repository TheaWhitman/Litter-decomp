```python
%%R -w 800 -h 1600
p = plot_bar(physeq, x="Rhizo", facet_grid=~Class, fill="ID")
p = p + geom_bar(aes(color=ID, fill=ID), stat="identity", position="stack")
p = p + theme(legend.position="none")
p = p + facet_wrap(~Class,ncol=2)
p <- p + theme(panel.background=element_blank())
p <- p + theme(panel.grid=element_blank(), axis.text=element_text(size=17, color="black"), axis.title=element_text(size=20,face="bold"))
p <- p + theme(legend.text=element_text(size=17, color="black"), legend.title=element_text(size=20,face="bold"))
p <- p + theme(strip.text=element_text(size=14, color="black"))
# Again, should be normalized
ggsave(filename = "CAZyBlast1.pdf", plot = p, width=8, height = 16)
p
```
