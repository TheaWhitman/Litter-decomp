```python
%%R
FastaStats.f<- function(x){
    Stats <- data.frame(ID=integer(), Day=integer(), Soil=integer(), nseqs=integer(), max=integer(), median=integer(), mean=integer(), min=integer(), stringsAsFactors=TRUE) 
    for (i in x){
        assembled.fa <- readDNAStringSet(i, format="fasta")
        ID <- sub("final_contigs(.*?).fa.*$", "\\1", i, perl = TRUE)
        Day <- Map[match(ID,Map$ID),]$Day
        Soil <- Map[match(ID,Map$ID),]$Soil
        nseqs<-length(assembled.fa)
        max<-max(nchar(assembled.fa))
        median<-median(nchar(assembled.fa))
        mean<-as.integer(mean(nchar(assembled.fa)))
        min<-min(nchar(assembled.fa))
        NewStats<-c(ID,Day,Soil,nseqs,max,median,mean,min)
        Stats[i,]<-NewStats
    }
    FastaStats<<-Stats
}
# This makes a function that loads the fasta file, measures its sequences, and outputs the data as "Stats", a global variable.
```

```python
%%R
FastaStats.f(filename)

# Using the function above on our list of files - outputs report of sequence data lengths
```

```python
%%R
FastaCounts.f<- function(x){
    Counts <- data.frame(ID=integer(), Day=integer(), Soil=character(), Lengths=integer(), stringsAsFactors=TRUE) 
    for (i in x){
        assembled.fa <- readDNAStringSet(i, format="fasta")
        IDs <- sub("final_contigs(.*?).fa.*$", "\\1", i, perl = TRUE)
        Lengths.df<-data.frame(ID=IDs, Day=Map[match(IDs,Map$ID),]$Day, Soil=Map[match(IDs,Map$ID),]$Soil, Lengths=c(as.numeric(sort(nchar(assembled.fa)))))
        # Lengths.df<-data.frame(ID=IDs, Lengths=c(as.numeric(sort(nchar(assembled.fa))[1:(nchar(assembled.fa)*0.99)])))
        # Option for taking only the top 99% of each set. But this doesn't work if one set has lots of weirdos.
        # Can just scale plots below instead of cutting data out here.
        Counts<-rbind(Counts,Lengths.df)
    }
    FastaCounts<<-Counts
}
# Creates the function that hopefully makes a list of the sequence lengths associated with their ID number, Day, & Soil
FastaCounts.f(filename)
dim(FastaCounts)
# Runs that function on the file lists, reports size of dataframe
```
