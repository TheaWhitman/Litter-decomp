```python
%%R
files = list.files(path = ".", pattern=c("uparse_blast_output_cazy_counts","*.txt"))
# Creates a list of all the uparse filenames that end with .txt in the current directory
Samples <- files
Samples <- gsub("(.+)_(.+)(.txt)","\\2",files)
Samples
# Gets the ID number (anything that comes between the last underscore and .txt (whatever is in the second set of parentheses))
```

```python
%%R
# Want to make a new table that has the CAZy ID, then the count for each sample, named as its sample ID
# Could modify to look up treatment, etc. based on sample ID, or group by them, etc. For now just ID.
CAZyTable.f <- function(x){
    CAZy <- data.frame()
    for (i in x) {
        Sample <- as.character(gsub("(.+)_(.+)(.txt)","\\2",i))
        CAZyt <- read.table(i, sep="\t", col.names=c("ID", Sample))
        CAZy<-join(CAZy, CAZyt, type="full")
    }
    CAZy[is.na(CAZy)] <- 0
    # Replaces NAs with 0's
    CAZyComb<<-CAZy
}
# It adds an X in front of the ID for now - can be fixed later
```

```python
%%R
CAZyTable.f(files)
CAZyComb.m <- data.matrix(CAZyComb[,2:dim(CAZyComb)[2]])
rownames(CAZyComb.m) <- paste0("OTU", 1:nrow(CAZyComb.m))
CAZyComb.m[1:6,1:2]
# This works on my test files where I rearranged the order, cut a value out, and made a giant value.
# Turns the data frame into a matrix so it can become a physeq file
# This is our fake OTU matrix
```

```python
%%R
getClass <- function (x){
    Class<-substr(x,1,2)
    return(Class)
}
taxmat = data.frame(CAZyComb[,1])
taxmat[,2] = getClass(CAZyComb[,1])
taxmat = as.matrix(taxmat)
rownames(taxmat) <- rownames(CAZyComb.m)
colnames(taxmat) <- c("ID","Class")
taxmat = taxmat[order(taxmat[,1]),]
taxmat
# Making a fake taxonomy matrix where the ID of the "OTUs" (i.e., CAZymes) are matched up with their real IDs
# Also orders it by Class
```

```python
%%R
sampledata <- sample_data(data.frame(read.csv("SampleData.txt")))
sampledata
# Importing fake sample data table
```

```python
%%R
CAZy = otu_table(CAZyComb.m, taxa_are_rows = TRUE)
IDs = tax_table(taxmat)
physeq <- phyloseq(CAZy, IDs, sampledata)
# Making temporary phyloseq object
```

```python
%%R
library("ape")
random_tree = rtree(ntaxa(physeq), rooted = TRUE, tip.label = taxa_names(physeq))
```
